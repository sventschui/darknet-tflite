# Build darknet

## Install dependencies & clone repo

```python
%%bash
# master or v5.1
DARKNET_BRANCH=master

apt-get install -y build-essential git libopencv-dev libprotobuf-dev

cd /content
rm -rf /content/darknet
git clone https://codeberg.org/CCodeRun/darknet.git
cd darknet
git checkout "${DARKNET_BRANCH}"
```

## Prepare build

```python
%%bash
rm -rf /content/darknet/build
mkdir /content/darknet/build
cd /content/darknet/build
cmake -DCMAKE_BUILD_TYPE=Release -DDARKNET_TRY_ONNX=True -DDARKNET_TRY_CUDA=$(if [[ -d '/usr/lib64-nvidia' ]]; then echo 'ON'; else echo "OFF"; fi) ..
```

## Recompile protobuf

```python
%%bash
cd /content/darknet/src-onnx
rm onnx.proto3
wget --no-clobber https://raw.githubusercontent.com/onnx/onnx/refs/heads/main/onnx/onnx.proto3
protoc --cpp_out=. onnx.proto3
```

## Compile darknet

```python
%cd /content/darknet/build
!make -j $(nproc)
```

```python
%cd /content/darknet/build
!src-cli/darknet version
```
