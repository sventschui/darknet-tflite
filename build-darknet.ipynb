# Build darknet

## Install dependencies & clone repo

```python
%%bash

set -e

apt-get install -y build-essential git libopencv-dev libprotobuf-dev

cd /content
rm -rf /content/darknet
git clone https://codeberg.org/CCodeRun/darknet.git
```

## Prepare build

```python
%%bash

set -e

rm -rf /content/darknet/build

mkdir /content/darknet/build

cd /content/darknet/build

cmake -DCMAKE_BUILD_TYPE=Release -DDARKNET_TRY_ONNX=True -DDARKNET_TRY_CUDA=$(if [[ -d '/usr/lib64-nvidia' ]]; then echo 'ON'; else echo "OFF"; fi) ..
```

## Compile darknet

```python
%%bash

set -e

cd /content/darknet/build

make -j $(nproc)

make package

sudo dpkg -i darknet*.deb
```
